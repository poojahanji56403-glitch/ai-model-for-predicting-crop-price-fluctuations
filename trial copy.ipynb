```python
def generate_dataset(n_samples=1000):

    crops = list(crop_base_price.keys())
    seasons = ["Kharif", "Rabi", "Zaid"]

    data = []

    for _ in range(n_samples):
        crop = random.choice(crops)
        season = random.choice(seasons)

        rainfall = np.random.uniform(200, 1500)
        temperature = np.random.uniform(15, 40)
        humidity = np.random.uniform(30, 90)
        soil_fertility = np.random.uniform(0.3, 1.0)
        fertilizer = np.random.uniform(50, 300)
        market_demand = np.random.uniform(0.5, 1.5)
        transport_cost = np.random.uniform(200, 2000)

        # Base price
        base_price = crop_base_price[crop]

        # Price calculation (logic-based)
        price = (
            base_price +
            rainfall * 0.5 +
            temperature * 10 +
            humidity * 5 +
            soil_fertility * 1000 +
            fertilizer * 2 +
            market_demand * 1500 -
            transport_cost * 0.3 +
            np.random.normal(0, 200)  # noise
        )

        data.append([
            crop, season, rainfall, temperature, humidity,
            soil_fertility, fertilizer, market_demand,
            transport_cost, round(price, 2)
        ])

    columns = [
        "Crop Type", "Season", "Rainfall (mm)", "Temperature (°C)",
        "Humidity (%)", "Soil Fertility Index", "Fertilizer Usage",
        "Market Demand Index", "Transportation Cost", "Crop Price"
    ]

    return pd.DataFrame(data, columns=columns)
```

```python
df = generate_dataset(4000)
df.to_csv("crop_price_dataset_100.csv", index=False)

df.head()
```
